```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
balloon_width=np.random.normal(15,1,200)
balloon_length=np.random.normal(20,2,200)
balloon_weigth=np.random.normal(0.01,0.5,200)
```

```python
melon_width=np.random.normal(20,1,200)
melon_length=np.random.normal(20,1,200)
melon_weigth=np.random.normal(2,1,200)
```

```python

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(balloon_width,balloon_length,balloon_weigth,color="blue",marker="p",label="balloon")
ax.scatter(melon_width,melon_length,melon_weigth,color="orange",marker="^",label="melon")
ax.set_zlabel("Weigth(kg)")

plt.xlabel("Width(cm)")
plt.ylabel("Length(cm)")
plt.title("Ballons & Melons")
plt.legend()
plt.show()
```
